```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px 
import seaborn as sns
df = pd.read_csv("songs_normalize.csv")
```

```python
df.groupby('year')["popularity","loudness"].mean().sort_values(by = "year")
```

```python
len(df["genre"].unique())
```

```python
plt.figure(figsize = (15,8))
plt.xticks(rotation = 90) # rotate x values
sns.countplot(x= "genre", data = df)
```

```python
fig = px.histogram(df, x='genre', title = 'Distribution of Genre')
fig.update_xaxes(categoryorder = "total ascending")
fig.update_xaxes(tickangle = -90)
fig.show()
```

```python
# good for quantitative data 
plt.figure(figsize = (15,8))
# sns.histplot(df["danceability"], binwidth = 0.05)
#sns.histplot(df["danceability"], bins = 30)
sns.histplot(df["danceability"], bins = 30, kde = True, stat = "density") # kernel density estimate, stat = density make y axis
# scale better
```

```python
fig = px.histogram(df, x = "danceability", title = "distribution of dance ability", nbins = 30)
fig.update_layout(bargap= 0.1)
fig.show()
```
